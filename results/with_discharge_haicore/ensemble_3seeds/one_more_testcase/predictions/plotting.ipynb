```python
# Plotting
import matplotlib.dates as mdates
#subset
start_date = '1994-06-23'
end_date = '1994-07-02'
with_disch_filtered = with_disch.loc[start_date:end_date,:]
without_disch_filtered = without_disch.loc[start_date:end_date,:]
neck_filtered = neck_1.loc[start_date:end_date,:]
fig, ax1 = plt.subplots(figsize=(9, 6))

# Plot the existing dataframes
ax1.plot(with_disch_filtered.index, with_disch_filtered['y_obs'], label='EOBS', marker='o')
ax1.plot(with_disch_filtered.index, with_disch_filtered['y_sim_mean'], label='with discharge', linestyle='--')
ax1.plot(neck_filtered.index, neck_filtered['total_precipitation_sum'], label='ERA5 Land', linestyle='-',color='grey')
ax1.plot(without_disch_filtered.index, without_disch_filtered['y_sim_mean'], label='without discharge', linestyle=':')


# Add labels and legend to the primary y-axis
ax1.set_xlabel('Historical flood at Neckarbischofheim (1994)', fontsize=14)
ax1.set_ylabel('Precipitation (mm/day)', fontsize=14)
#ax1.set_title(f'Catchment: Elsenz Schwarzbach from {start_date} to {end_date}')
ax1.legend(loc='center right', fontsize=12)
ax1.set_ylim([0, 100])
ax1.set_yticks([0,20,40,60])
# Formatting tick labels
ax1.tick_params(axis='both', labelsize=12)
# Create a secondary y-axis
ax2 = ax1.twinx()
ax2.plot(neck_filtered.index, neck_filtered['streamflow'], label='Streamflow', color='red', linestyle='-', linewidth=2)
ax2.fill_between(neck_filtered.index, neck_filtered['streamflow'], color='red', alpha=0.3)
ax2.scatter(pd.to_datetime('1994-06-27'), neck_filtered.loc[pd.to_datetime('1994-06-27'),'streamflow'], color='red', s=75,marker='x', zorder=5)
ax2.set_ylabel('Streamflow (mm/day)', loc='top', fontsize=14)
ax2.legend(loc='upper right', fontsize=12)
ax2.set_ylim([0, 300])
ax2.set_yticks([0,20,40,60,80,100])
ax2.tick_params(axis='both', labelsize=12)
ax2.invert_yaxis()

# Format the date labels on the x-axis
date_format = mdates.DateFormatter('%b %d')
ax1.xaxis.set_major_formatter(date_format)
#fig.autofmt_xdate()  # Rotate date labels
# Formatting the plot
#plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig("thesis.jpeg",dpi=1200)
# Show the plot
plt.show()
```
