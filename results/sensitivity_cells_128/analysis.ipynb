```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
nse_5  = pd.read_csv(r"C:\Users\ashis\Github\lstm_backward\results\sensitivity_cells_5\NSE.csv")
nse_16  = pd.read_csv(r"C:\Users\ashis\Github\lstm_backward\results\sensitivity_cells_16\NSE.csv")
nse_32  = pd.read_csv(r"C:\Users\ashis\Github\lstm_backward\results\sensitivity_cells_32\NSE.csv")
nse_64  = pd.read_csv(r"C:\Users\ashis\Github\lstm_backward\results\with_discharge\NSE.csv")
nse_128  = pd.read_csv(r"C:\Users\ashis\Github\lstm_backward\results\sensitivity_cells_128\NSE.csv")
```

```python
# Add a new column to each DataFrame indicating the DataFrame source
nse_5['Source'] = 'nse_5'
nse_16['Source'] = 'nse_16'
nse_32['Source'] = 'nse_32'
nse_64['Source'] = 'nse_64'
nse_128['Source'] = 'nse_128'
```

```python
# Combine all DataFrames into a single DataFrame
combined_df = pd.concat([nse_5, nse_16,nse_32,nse_64,nse_128])
combined_df = combined_df[combined_df["NSE"] >=0]
```

```python
# Plot the violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Source', y='NSE', data=combined_df)
plt.xlabel('Hidden Cells')
plt.ylabel('NSE')
plt.show()
```
