```python
file_path = r'U:\02_Software\Github\lstm_backward\results\with_discharge\NSE.csv'
nse_disch_42 = pd.read_csv(file_path)
file_path = r'U:\02_Software\Github\lstm_backward\results\with_discharge_1500\NSE.csv'
nse_disch_1500 = pd.read_csv(file_path)
file_path = r'U:\02_Software\Github\lstm_backward\results\with_discharge_8\NSE.csv'
nse_disch_8 = pd.read_csv(file_path)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming the relevant NSE column is named 'NSE'
# You may need to adjust 'NSE' to match your actual column name
nse_disch_42['Discharge'] = '42'
nse_disch_1500['Discharge'] = '1500'
nse_disch_8['Discharge'] = '8'

# Combine the dataframes
combined_nse = pd.concat([nse_disch_42[['NSE', 'Discharge']],
                           nse_disch_1500[['NSE', 'Discharge']],
                           nse_disch_8[['NSE', 'Discharge']]])

# Create the violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Discharge', y='NSE', data=combined_nse)
plt.title('Violin Plot of NSE Across Different Discharge Conditions')
plt.xlabel('Discharge Condition')
plt.ylabel('NSE Value')
plt.show()
```
